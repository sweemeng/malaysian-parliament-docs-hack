First let's download a utilities

```python
import data_extractor
```

It is just set of utilities to convert the whole list of pdf to text

```python
documents = data_extractor.extract_text(path="docs/results/")
```

Now we need to tokenize each documents. On hindsight, might be a good idea to tokenize in library. 


We need to split tech into list. So we need to tokenize it. We will use nltk for this

```python
import nltk
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'\w+') # Spoiler this is to extract word without punctuation
```

Now we tokenize the document. In list! 

```python
docs = []
```

```python
for document in documents:
    docs.append(tokenizer.tokenize(document))
```

```python
docs
```

```python
docs[0]
```

Now let's turn this into a corpus gensim can use

```python
from gensim import corpora
```

Create a dictionary and save it

```python
dictionary = corpora.Dictionary(docs)
```

```python
dictionary.save("malaysian_parl.dict")
```

```python
print(dictionary)
```

```python
dictionary.token2id
```

```python
dictionary.doc2bow(["Kementerian","Pertahanan"])
```

```python
corpus = [dictionary.doc2bow(doc) for doc in docs]
```

```python
corpora.MmCorpus.serialize("malaysian_parl.mm", corpus)
```
